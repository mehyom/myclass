```python
# 1-A
import nltk
nltk.download('gutenberg')
from nltk.corpus import gutenberg
nltk.download('punkt')
hamlet = 'shakespeare-hamlet.txt'
num_chars = len(gutenberg.raw(hamlet))
num_words = len(gutenberg.words(hamlet))
num_sents = len(gutenberg.sents(hamlet))
print(round(num_chars/num_words),round(num_words/num_sents))
```

```python
# 1-B
import nltk
nltk.download('gutenberg')
from nltk.corpus import gutenberg
nltk.download('punkt')
hamlet = 'shakespeare-hamlet.txt'
hamlet_word = gutenberg.words(hamlet)
longest_len = max(len(s) for s in hamlet_word)
print(longest_len, [w for w in hamlet_word if len(w) == longest_len])
```

```python
# 1-C
import nltk
nltk.download('gutenberg')
from nltk.corpus import gutenberg
nltk.download('punkt')
hamlet = 'shakespeare-hamlet.txt'
num_vocab = len(set(w.lower() for w in gutenberg.words(hamlet)))
print(round(len(gutenberg.words(hamlet))/num_vocab))
```

```python
# 1-D
import nltk
nltk.download('gutenberg')
from nltk.corpus import gutenberg
nltk.download('punkt')
hamlet = 'shakespeare-hamlet.txt'
hamlet_word = gutenberg.words(hamlet)
nltk.download('names')
male_names = nltk.corpus.names.words('male.txt')
print(set([w for w in hamlet_word[:1000] if w in male_names]))
```

```python
# 1-E
import nltk
nltk.download('gutenberg')
from nltk.corpus import gutenberg
nltk.download('punkt')
hamlet = 'shakespeare-hamlet.txt'
hamlet_word = gutenberg.words(hamlet)
import re
len(set([w for w in hamlet_word if re.search('^..i[A-Za-z]{7}$',w)]))
```

```python
import nltk
nltk.download('brown')
nltk.download('universal_tagset')
from nltk.corpus import brown
```

```python
brown_news_tagged = brown.tagged_words(categories='news',tagset='universal')
```

```python
tag_fd = nltk.FreqDist(tag for (word, tag) in brown_news_tagged)
```

```python
pos_list = []
for pos in tag_fd.most_common():
    pos_list.append(pos[0])
print(pos_list[pos_list.index('ADJ')+1])
```
