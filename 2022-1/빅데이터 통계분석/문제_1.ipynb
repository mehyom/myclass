```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from(['A','B','C','D','E','F','G','H','I','J'])
G.add_edges_from([('A','C'),('A','D'),('A','E'),('B','D'),('B','E'),('D','E'),('D','F'),('D','G'),('E','G'),('F','G'),('G','H'),('G','I'),('G','J')])
nx.draw(G, with_labels=True, node_size=400, node_color='skyblue')
plt.show()
```

```python
degree_freq = nx.degree_histogram(G)
x = range(len(degree_freq))
plt.bar(x, degree_freq)
plt.show()
print(nx.degree(G))
```

```python
print('Degree Centrality:', nx.degree_centrality(G))
```
