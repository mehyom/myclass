```python
import pandas as pd

df = pd.read_csv("/content/smoking.csv")
```

```python
x = df.drop(['ID', 'smoking'], axis=1)
y = df['smoking']
```

```python
x = x.replace('Y', 1)
x = x.replace('N', 0)
```

```python
x
```

```python
x.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,5))
sns.kdeplot(data=x.groupby('gender').get_group('F')['height(cm)'], color='red', shade=True)
sns.kdeplot(data=x.groupby('gender').get_group('M')['height(cm)'], color='blue', shade=True)
plt.show()
```

```python
from scipy import stats

t_stat, p_val = stats.ttest_ind(x.groupby('gender').get_group('F')[x.columns],
                                x.groupby('gender').get_group('M')[x.columns],
                                equal_var=False)
```

```python
t_stat
```

```python
for i, j in zip(x.columns, p_val):
    print(i, ': ', round(j, 4))
```

```python
discrete = ['gender', 'oral', 'dental caries', 'tartar']
continuous = ['age', 'height(cm)', 'weight(kg)', 'waist(cm)',
              'eyesight(left)', 'eyesight(right)', 'hearing(left)', 'hearing(right)',
              'systolic', 'relaxation', 'fasting blood sugar', 'Cholesterol',
              'triglyceride', 'HDL', 'LDL', 'hemoglobin', 'Urine protein',
              'serum creatinine', 'AST', 'ALT', 'Gtp']
```

```python
for c in continuous:
    print(c, ': ', round(stats.pointbiserialr(x[c], y)[0], 4))
```

```python
for c in discrete:
    print('\n[', c, ']\n\n', x[c].value_counts(), '\n')
    print(pd.crosstab(x[c], y), '\n\n')
```
